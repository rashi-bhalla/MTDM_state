```clojure
(import '[java.awt Color]
        '[com.twosigma.beakerx.chart.legend LegendPosition LegendLayout]
        '[com.twosigma.beakerx.chart.xychart Plot CombinedPlot]
        '[com.twosigma.beakerx.chart.xychart.plotitem Points ShapeType])

(def colors [(Color/decode "#ff96a8") (Color/decode "#f55774")])
(def outline-colors [(Color/black) (Color/black) ])
(def shapes [ShapeType/CIRCLE ShapeType/TRIANGLE])

```

```clojure
(let [no-plot (doto (Plot.)
                   (.setLegendPosition LegendPosition/TOP)
                   (.setLegendLayout LegendLayout/HORIZONTAL)
                   (.setYLabel "No of combinations")
                   (.setXBound [700 8000])
                   (.setYBound [10 40]))
      time-gen-plot (doto (Plot.)
                     (.setLegendPosition LegendPosition/TOP)
                     (.setLegendLayout LegendLayout/HORIZONTAL)
                     (.setYLabel "Resource time")
                     (.setXBound [700 8000])
                     (.setYBound [0.0001 2.2]))
      
      plot (doto (CombinedPlot.)
               (.add no-plot)
               (.add time-gen-plot)
               (.setInitWidth 450)
               (.setInitHeight 750)
               (.setXLabel "Records"))]
               
               (.add no-plot
                  (doto (Points.)
                      (.setX (keys (first (vals results))))
                      (.setY (vals (first (vals results))))
                      (.setShape ShapeType/CIRCLE)
                      (.setColor (Color/decode "#ff96a8"))
                      (.setOutlineColor (Color/black))
                      (.setSize 10)
                      (.setDisplayName "Combinations")
                      #_(.setFill false)
                
                 ) )
                 
          (.add time-gen-plot
            (doto (Points.)
                 (.setX (keys (nth (first (second (vals results))) 0)))
                     (.setY (vals (nth (first (second (vals results))) 0)))
                   (.setShape ShapeType/CIRCLE)
                      (.setColor (Color/decode "#ff96a8"))
                     (.setOutlineColor (Color/black))
                      (.setSize 10)
                     (.setDisplayName "Generation")
                      #_(.setFill false)
                
                 ) )
                 
                 (.add time-gen-plot
            (doto (Points.)
                 (.setX (keys (nth (first (second (vals results))) 1)))
                     (.setY (vals (nth (first (second (vals results))) 1)))
                   (.setShape ShapeType/TRIANGLE)
                      (.setColor (Color/decode "#f55774"))
                     (.setOutlineColor (Color/black))
                      (.setSize 10)
                      #_(.setFill false)
                 (.setDisplayName "Parsing")
                 ) )
                 
               
     ;;    (println  (keys (nth (first (second (vals results))) 0)))
        
       plot )   
               
               
```
